# Nettoyage des donn√©es et premiers analyses

```python
import pandas as pd
import numpy as np
```

```python
# Fais apparaitre l'ensemble des informations
pd.set_option("display.max_rows", 5000)
pd.set_option("display.max_column", 70)
pd.set_option("display.max_colwidth", 100)
```

```python
# 1- Import des donn√©es
data = pd.read_csv('../Datasets/Exoplanets.csv', skiprows=96)
data.head()
```

```python
# 2- Extraction de la liste du nom des colonnes
data.columns
print("Voici la liste des colonnes :", data.columns.tolist())
print("Nombre de colonnes :", len(data.columns))
```

```python
# 3- Taille du dataset
data.shape
print("Le dataset contient", data.shape[0], "lignes et", data.shape[1], "colonnes !")
```

```python
# 4- Filtrer les colonnes avec des valeurs utiles
columns = [
    'pl_name', 'hostname', 'sy_snum', 'sy_pnum','discoverymethod','disc_year','pl_controv_flag',
    'pl_orbper', 'pl_orbsmax', 'pl_rade', 'pl_radj', 'pl_bmasse', 'pl_bmassj', 'pl_insol', 'pl_eqt', 'st_teff', 'st_rad', 'st_mass', 'sy_dist'
]

data = data[columns]
print("Ici nous gardons uniquement les colonnes qui sont selon nous utiles pour notre projet :)")
data.head()
```

```python
# 5- Renommer les colonnes 
data = data.rename(columns={
    'pl_name': 'Planet Name',
    'hostname': 'Host Name',
    'sy_snum': 'Number of Stars in System',
    'sy_pnum': 'Number of Planets in System',
    'discoverymethod': 'Discovery Method',
    'disc_year': 'Discovery Year',
    'pl_controv_flag': 'Controversial Flag',
    'pl_orbper': 'Orbital Period',
    'pl_orbsmax': 'Semi-Major Axis',
    'pl_rade': 'Planet Earth Radius',
    'pl_radj': 'Planet Jupiter Radius',
    'pl_bmasse': 'Planet Earth Mass',
    'pl_bmassj': 'Planet Jupiter Mass',
    'pl_insol': 'Insolation Flux',
    'pl_eqt': 'Equilibrium Temperature',
    'st_teff': 'Stellar Effective Temperature',
    'st_rad': 'Stellar Radius',
    'st_mass': 'Stellar Mass',
    'sy_dist': 'Distance'
})

print("Renommage des colonnes pour plus de clart√©")
data.head()
```

```python
# 6- Afficher taux de remplissage de chaque colonne
def print_fill_rate(data):
    nb_rows = data.shape[0]
    fill_rate = data.notna().sum() / nb_rows * 100
    return fill_rate.rename('Fill Rate (%)')

fill_rate = print_fill_rate(data)

print("Voici le pourcentage de remplissage de chaque colonne :")
fill_rate 
```

```python
# 7- Rep√©rage des valeurs nulles
print("Voici le nombre de valeurs nulles par colonne:")
print(data.isnull().sum())

# Les valeurs nulles dans les donn√©es d'exoplan√®tes sont g√©n√©ralement dues √† des limitations techniques plut√¥t qu'√† des oublis
# Par exemple, certaines m√©thodes de d√©tection ne permettent pas de mesurer tous les param√®tres (la m√©thode de transit donne le rayon mais pas toujours la masse)
```

```python
# 8- Combien y a-t-il de plan√®tes diff√©rentes d√©couvertes ?
unique_planets = data['Planet Name'].nunique()
print(unique_planets, "plan√®tes diff√©rentes d√©couvertes.")
```

```python
# 9- Combien y a-t-il de syst√®mes doubles ?
double_systems = len(data[data['Number of Stars in System'] == 2]['Host Name'].unique())
print(double_systems, "syst√®mes doubles d√©tect√©s.")
```

```python
# 10- Combien y a-t-il de d√©couvertes controvers√©es ? On v√©rifie aussi les doublons ! Eh oui, il y en a !
controversial_discovery = data[data['Controversial Flag'] == 1]['Planet Name'].nunique()
print(f"Nombre de plan√®tes controvers√©es : {controversial_discovery}")
```

```python
# 11- En quelle ann√©e a-t-on d√©couvert le plus d‚Äôexoplan√®tes ?
most_discoveries_year = data['Discovery Year'].value_counts().sort_values(ascending=False)
most_discoveries_year.head()
```

```python
# 12- Voir si y'a des doublons
check_duplicate = data["Planet Name"].duplicated().value_counts()
data = data.drop_duplicates(subset=["Planet Name"])

print("Voici le nombre de ligne restantes apr√®s suppression des doublons :", data.shape[0])
```

```python
# Enregistre le dataset nettoy√©
data.to_csv('../Datasets/Exoplanets_cleaned.csv', index=False)
```
